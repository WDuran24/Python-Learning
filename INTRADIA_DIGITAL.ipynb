```python
# Definir una función para validar el formato del RUT
def validar_rut(rut):
    patron = r"^\d{7,8}-[0-9Kk]$"  # Ejemplo: 15478996-0 o 15478996-K
    return bool(re.match(patron, str(rut)))

# Revisar cada fila y buscar RUT en otras columnas
columnas_a_revisar = [col for col in f_blip_ha.columns if col != "rut"]  # Todas las columnas excepto "RUT"

for index, row in f_blip_ha.iterrows():
    # Validar si el campo RUT es válido
    if not validar_rut(row["rut"]):
        # Buscar un RUT válido en otras columnas
        for col in columnas_a_revisar:
            if validar_rut(row[col]):
                f_blip_ha.at[index, "rut"] = row[col]  # Mover el RUT al campo correcto
                break  # Dejar de buscar una vez encontrado
```

```python
def calcular_semana_lunes(fecha):
    if pd.isnull(fecha):
        return None
    # Primer día del mes
    primer_dia_mes = fecha.replace(day=1)
    # Determinar el lunes de la primera semana del mes
    primer_lunes = primer_dia_mes + timedelta(days=(7 - primer_dia_mes.weekday()) % 7)
    if primer_lunes.month != fecha.month:
        primer_lunes -= timedelta(days=7)  # Ajustar si el primer lunes cae en el mes siguiente
    # Calcular el número de semana
    delta = (fecha - primer_lunes).days
    return delta // 7 + 1 if delta >= 0 else 1

f_blip_ha['Semana'] = f_blip_ha['Fecha'].apply(calcular_semana_lunes)
```

```python
def agregar_campo_def_ibr(f_blip_ha):
    """
    Agrega un campo llamado 'Def_IBR' a la base de datos f_blip_ha.
    """
    # Asegurarse de que no haya valores nulos para evitar errores de búsqueda
    f_blip_ha.fillna("", inplace=True)
    
    # Función para verificar la condición
    def verificar_def_ibr(row):
        if "ivr_def5" in row['Campaign'].lower():
            return 1
        elif any("ivr_def5" in str(value).lower() for value in row):
            return 1
        else:
            return 0
    
    # Crear el nuevo campo aplicando la función a cada fila
    f_blip_ha['Def_IBR'] = f_blip_ha.apply(verificar_def_ibr, axis=1)
agregar_campo_def_ibr(f_blip_ha)  # Modifica directamente f_blip_ha
```

```python
# Lista de columnas a conservar
columnas_deseadas = [
    "SequentialId", "CustomerIdentity", "AgentIdentity", 
    "Status", "StorageDate", "Team", "AgentName", "AgentEmail"
]

# Filtrar el DataFrame
blip_ha_FCR = blip_ha[columnas_deseadas]
```

```python
# Crear el nuevo campo 'Fecha' extrayendo solo la parte de la fecha del campo 'StorageDate'
blip_ha_FCR['Fecha'] = pd.to_datetime(blip_ha_FCR['StorageDate']).dt.date
```

```python
# Asegurar que la columna 'Fecha' es de tipo datetime
blip_ha_FCR['Fecha'] = pd.to_datetime(blip_ha_FCR['Fecha'])

# Convertir 'team' a minúsculas, reemplazando NaN con cadena vacía
blip_ha_FCR['team_lower'] = blip_ha_FCR['Team'].astype(str).str.lower().fillna("")

# Función para calcular Flag_Konecta
def calcular_flag_konecta(row):
    if ("comercial_fijo_def" in row['team_lower'] or "tecnico_fijo_def" in row['team_lower']) and row['Fecha'] >= pd.Timestamp(2024, 11, 28):
        return 1
    elif "op_def" in row['team_lower'] and row['Fecha'] >= pd.Timestamp(2024, 12, 11):
        return 1
    elif "tecnicofinix_def" in row['team_lower'] and row['Fecha'] >= pd.Timestamp(2024, 12, 16):
        return 1
    elif "soporte_ott" in row['team_lower'] and row['Fecha'] >= pd.Timestamp(2024, 12, 19):
        return 1
    elif "op" in row['team_lower'] and row['Fecha'] >= pd.Timestamp(2024, 12, 23):
        return 1
    elif any(sub in row['team_lower'] for sub in [
        "transf desde movil", "masivos", "venta_ott", "transf desde ventas",
        "default", "transferencia directa", "consulta_boleta", "ott_max",
        "dbox smart", "ftth/iptv"
    ]) and row['Fecha'] >= pd.Timestamp(2025, 1, 15):
        return 1
    else:
        return 0

# Aplicar la función
blip_ha_FCR['Flag_Konecta'] = blip_ha_FCR.apply(calcular_flag_konecta, axis=1)

# Eliminar la columna auxiliar
blip_ha_FCR.drop(columns=['team_lower'], inplace=True)
```

```python
# Define the file path
archivo = r"C:\Users\wduran\OneDrive - CLAROCHILE\Digital Team\Dotaciones\VTR\Dotación_VTR_20250228.xlsx"
# Read the specific sheet named 'CONSOLIDADO' from the Excel file
dotacion = pd.read_excel(archivo, sheet_name='Dotacion')  # Specify the sheet name
```

```python
# Crear la nueva columna MAIL que extrae la parte izquierda del arroba
dotacion['MAIL'] = dotacion['CORREO_ELECTRONICO'].str.split('@').str[0]
```

```python
# Crear la nueva columna MAIL que extrae la parte izquierda del arroba
blip_ha_FCR['MAIL'] = blip_ha_FCR['AgentIdentity'].str.split('%').str[0]
```

```python
blip_ha_FCR = blip_ha_FCR.merge(dotacion[['MAIL', 'ID_ANDES','EMPRESA_PAIS','PLATAFORMA','SUPERVISOR']], on='MAIL', how='left')
```

```python
# Eliminar el campo 'mail'
blip_ha_FCR.drop(columns=['MAIL'], inplace=True)
```

```python
blip_ha_FCRDUPLI=blip_ha_FCR
```

```python
# Reducir tamaño de las columnas antes del merge
blip_ha_FCRDUPLI_reduced = blip_ha_FCRDUPLI[['CustomerIdentity', 'Fecha','Status','SequentialId']].copy()

# Realizar el join con la menor cantidad de datos posible
merged_df = blip_ha_FCR.merge(blip_ha_FCRDUPLI_reduced, on='CustomerIdentity', how='left', suffixes=('_FCR', '_FCRDUPLI'))

# Convertir a datetime para evitar errores en la resta de fechas
merged_df['Fecha_FCRDUPLI'] = pd.to_datetime(merged_df['Fecha_FCRDUPLI'])
merged_df['Fecha_FCR'] = pd.to_datetime(merged_df['Fecha_FCR'])

# Calcular la diferencia en días solo para valores no nulos
mask = merged_df['Fecha_FCRDUPLI'].notna()
merged_df.loc[mask, 'FCR'] = (merged_df.loc[mask, 'Fecha_FCRDUPLI'] - merged_df.loc[mask, 'Fecha_FCR']).dt.days

# Eliminar filas donde FCR < 1
merged_df = merged_df.loc[merged_df['FCR'] >= 1].copy()

# Filtrar solo registros donde status_fcr sea "closedattendant"
merged_df = merged_df[merged_df['Status_FCR'] == 'ClosedAttendant'].copy()
```

```python
#duplicamos la base 
blip_ha_FCRDUPLI=blip_ha_FCR
```

```python
merged_df.drop_duplicates(inplace=True)# Asegurar que UNICOS sea un entero
merged_df['FCR'] = merged_df['FCR'].astype(int)
```

```python
# Guardar el DataFrame como archivo CSV
merged_df.to_csv(r"C:\Users\wduran\OneDrive - CLAROCHILE\Digital Team\FCR\FCR Historico.csv", index=False)
```
