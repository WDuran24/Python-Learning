```python
# Definir una funci√≥n para validar el formato del RUT
def validar_rut(rut):
    patron = r"^\d{7,8}-[0-9Kk]$"  # Ejemplo: 15478996-0 o 15478996-K
    return bool(re.match(patron, str(rut)))

# Revisar cada fila y buscar RUT en otras columnas
columnas_a_revisar = [col for col in f_blip_ha.columns if col != "rut"]  # Todas las columnas excepto "RUT"

for index, row in f_blip_ha.iterrows():
    # Validar si el campo RUT es v√°lido
    if not validar_rut(row["rut"]):
        # Buscar un RUT v√°lido en otras columnas
        for col in columnas_a_revisar:
            if validar_rut(row[col]):
                f_blip_ha.at[index, "rut"] = row[col]  # Mover el RUT al campo correcto
                break  # Dejar de buscar una vez encontrado
```

```python
def calcular_semana_lunes(fecha):
    if pd.isnull(fecha):
        return None
    # Primer d√≠a del mes
    primer_dia_mes = fecha.replace(day=1)
    # Determinar el lunes de la primera semana del mes
    primer_lunes = primer_dia_mes + timedelta(days=(7 - primer_dia_mes.weekday()) % 7)
    if primer_lunes.month != fecha.month:
        primer_lunes -= timedelta(days=7)  # Ajustar si el primer lunes cae en el mes siguiente
    # Calcular el n√∫mero de semana
    delta = (fecha - primer_lunes).days
    return delta // 7 + 1 if delta >= 0 else 1

f_blip_ha['Semana'] = f_blip_ha['Fecha'].apply(calcular_semana_lunes)
```

```python
def agregar_campo_def_ibr(f_blip_ha):
    """
    Agrega un campo llamado 'Def_IBR' a la base de datos f_blip_ha.
    """
    # Asegurarse de que no haya valores nulos para evitar errores de b√∫squeda
    f_blip_ha.fillna("", inplace=True)
    
    # Funci√≥n para verificar la condici√≥n
    def verificar_def_ibr(row):
        if "ivr_def5" in str(row.get('Campaign', '')).lower():
            return 1
        elif any("ivr_def5" in str(value).lower() for value in row):
            return 1
        else:
            return 0
    
    # Crear el nuevo campo aplicando la funci√≥n a cada fila
    f_blip_ha['Def_IBR'] = f_blip_ha.apply(verificar_def_ibr, axis=1)
agregar_campo_def_ibr(f_blip_ha)  # Modifica directamente f_blip_ha
```

```python
equipos_claro = [
    "Cl_Tecnica_Cierre_Asegurado",
    "Cuenta Exacta Comercial",
    "Cuenta Exacta Tecnico",
    "Gesti√≥n",
    "Instalaci√≥n Pendiente",
    "Postpago Comercial",
    "Postpago Tecnico",
    "trans_d_movil",
    "trans_d_ventas",
    "CE Otras Consultas",
    "Transferencia Directa",
    "Prepago Tecnico",
    "Tecnico Hogar",
    "Comercial Hogar",
    "visita_tecnica",
    "Abuse",
    "Bienvenida_OUT",
    "masivo",
    "Transferencia directa",   # (distinta capitalizaci√≥n)
    "Babysitting Hogar - Dia 3",
    "Babysitting Hogar - Dia 14",
    "clarobot",
]

# normalizamos a min√∫sculas para evitar problemas de may√∫sculas/min√∫sculas,
# y quitamos espacios en blanco sobrantes al comienzo / final
equipos_claro_lower = [e.lower() for e in equipos_claro]

# Crea la columna MARCA
merged_df["MARCA"] = np.where(
    merged_df["Team"].str.strip().str.lower().isin(equipos_claro_lower),
    "Claro",     # cuando hay match
    "VTR"        # cuando NO hay match
)
```

```python
# --- 1) Definir el diccionario (OrderedDict para respetar el orden/prioridad) ---
category_map = OrderedDict([
    ("Conectividad",       ["sin internet", "intermitencia", "velocidad", "falla de servicio",
                            "sin tono", "no recibe", "dx de servicio"]),
    ("Television_Video",   ["sin se√±al tv", "canal", "vod", "replay", "control remoto",
                            "dbox", "streaming", "pixeleo", "grilla"]),
    ("Facturacion_Cobros", ["factur", "boleta", "cobro", "repo", "deuda", "saldo",
                            "descuento", "ajuste", "promo no aplicado"]),
    ("Pagos_Prorroga",     ["pago no aplicado", "pat-pac", "prorroga", "inhibicion",
                            "opcion de pago", "consulta saldo"]),
    ("Instalacion_Visita", ["visita", "instalaci", "autoinstal", "delivery", "entrega",
                            "traslado", "escalado", "mesa de ayuda"]),
    ("Plan_Promocion",     ["cambio de plan", "oferta", "cotiz", "actuali", "promo",
                            "venta", "cambio de ciclo", "modificaciones contrato",
                            "cambio de titular"]),
    ("Portabilidad_Equipo",["portabilidad", "cambio de equipo", "agregar servicio"]),
    ("SVA_Premium",        ["sva", "canal premium", "baja de sva"]),
    ("Digital_App",        ["sucursal virtual", "app", "videollamada", "boton rojo",
                            "1_streaming", "2_streaming", "trans_"]),
    ("Otros",              [])  # catch‚Äëall
])

# --- 2) Funci√≥n utilitaria para normalizar texto (quitar tildes y pasar a min√∫sculas) ---
def _normalize(text):
    if not isinstance(text, str):
        return ""
    text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("ascii")
    return text.lower()

# --- 3) Funci√≥n que asigna la categor√≠a a un registro ---
def categorize_tags(tag_cell):
    """
    tag_cell puede ser:
      ‚Ä¢ una lista real (['Sin internet', 'Sin se√±al TV'])
      ‚Ä¢ una string tipo "['Sin internet', 'Sin se√±al TV']"
      ‚Ä¢ una string simple ('Sin internet')
    """
    # Convertir a lista de strings
    if isinstance(tag_cell, list):
        tags = tag_cell
    else:
        try:
            tags = ast.literal_eval(tag_cell)  # intenta interpretar como lista en texto
            if not isinstance(tags, list):
                tags = [tag_cell]
        except Exception:
            tags = [tag_cell]

    # Normalizar y buscar
    norm_tags = [_normalize(t) for t in tags]

    for cat, keywords in category_map.items():
        if not keywords:          # "Otros"
            continue
        for kw in keywords:
            if any(kw in t for t in norm_tags):
                return cat
    return "Otros"

# --- 4) Aplicar a tu DataFrame ---
merged_df["Motivo_FCR"] = merged_df["Tags"].apply(categorize_tags)
```
